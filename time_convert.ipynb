```python
import numpy as np
import pandas as pd

df = pd.read_csv('./email_message_timestamp_merge.csv', encoding='utf-8-sig')
df = df.sort_values(by='new_time', ascending=True)

new_time = df['new_time'].values.tolist()

new_time
```

```python
# 구간 설정
epoch_list = [1732528800, 1732615200, 1732701600, 1732788000, 1732874400, 1733133600, 1733220000, 1733306400, 1733392800, 1733479200]
epoch_full_list = []

for i in range(len(epoch_list)):
    for j in range(5):
        epoch_full_list.append(epoch_list[i] + 7200 * j)
    if epoch_list[i] == 1732788000 or epoch_list[i] == 1733133600 or epoch_list[i] == 1733479200:
        epoch_full_list.append(epoch_list[i] + 7200 * 5)

epoch_full_list = sorted(epoch_full_list)
epoch_full_list
```

```python
# 구간 별 마지막 index
idx_lasttime = [83, 112, 143, 186, 186, 254, 290, 347, 400, 476, 518, 564, 602, 678, 708, 732, 766, 773, 832, 887, 923, 953, 1007, 1019, 1081, 1125, 1150, 1186, 1210, 1260, 1286, 1313, 1354, 1378, 1405, 1425, 1476, 1568, 1581, 1609, 1643, 1661, 1688, 1695]

n_values = []

for idx in idx_lasttime:
    remainder = new_time[idx] % 3600
    if remainder != 0:
        n_max = 7200 // remainder
    else:
        n_max = float('inf')
    n_values.append(n_max)

n_values
```

```python
i = 0
final_time = []

for k, idx in enumerate(idx_lasttime):
    remains = np.array(new_time[i:idx+1])
    remainder = remains % 3600

    for j in range(len(remains)):
        final_time.append(remains[j] + remainder[j] * (n_values[k] - 1))

    i = idx + 1

for j in range(i, len(new_time)):
    final_time.append(new_time[j])

len(final_time)
```

```python
df.insert(29, 'final_time', final_time)

df.to_csv('./email_message_timestamp_merge_final.csv', encoding='utf-8-sig', index=False)
```
